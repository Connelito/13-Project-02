```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
fuel_econ_df = pd.read_csv('fuel_economy.csv')
```

```python
fuel_econ_df.head()
```

```python
clean_fuel_econ_df = fuel_econ_df[['year',
                                   'make',
                                   'model',
                                   'city08',
                                   'highway08',
                                   'comb08',
                                   'fuelCost08',
                                   'cylinders',
                                   'trany',
                                   'drive']]
```

```python
clean_fuel_econ_df.head()
```

```python
print(len(clean_fuel_econ_df))
```

```python
clean_fuel_econ_rename = clean_fuel_econ_df.rename(columns={"year":"Year",
                                                            "make":"Make",
                                                            "model":"Model",
                                                            "city08":"City MPG",
                                                            "highway08":"Highway MPG",
                                                            "comb08":"Combined MPG",
                                                            "fuelCost08":"Annual Fuel Cost",
                                                            "cylinders":"Engine Cylinders",
                                                            "trany":"Transmission",
                                                            "drive":"Drive Axle Type"})
```

```python
clean_fuel_econ_rename.head()
```

```python
clean_fuel_econ_rename.to_csv("clean_fuel_economy.csv", 
                  encoding="utf-8", index=False, header=True)
```
