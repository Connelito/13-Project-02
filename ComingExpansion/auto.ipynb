```python
car_urls ='https://www.kbb.com/cars-for-sale/vehicledetails.xhtml?0=cars&listingId=592484557'
browser.visit(car_urls)

vehicle_details=[]  

html = browser.html
soup = bs(html, 'lxml')

for car in car_urls:

    try:

        title=soup.find('h1').text
        price=soup.find('span', class_='first-price').text

        details=soup.find('li', class_='list-bordered')

        miles=soup.find('div', class_='col-xs-10').text
        mileage=miles.replace('miles','')

        exterior=details.find_next_sibling('li')
        exterior_color=exterior.find('div', class_='col-xs-10').text

        interior=exterior.find_next_sibling('li')
        interior_color=interior.find('div', class_='col-xs-10').text

        eng=interior.find_next_sibling('li')
        engine=eng.find('div', class_='col-xs-10').text

        trans=eng.find_next_sibling('li')
        transmission=trans.find('div', class_='col-xs-10').text

        whl=trans.find_next_sibling('li')
        wheels=whl.find('div', class_='col-xs-10').text

        cty=whl.find_next_sibling('li')
        mpg=cty.find('div', class_='col-xs-10').text

        xpath = '/html/body/div[1]/div[2]/div/div/div[2]/div/div/div[1]/span'
        aa =browser.find_by_xpath(xpath)
        vin=aa[0].text

    except:pass

vehicle_details.append({
    'title': title,
    'price': price,
    'mileage': mileage,
    'exterior_color': exterior_color,
    'interior_color': interior_color,
    'engine': engine,
    'transmission': transmission,
    'wheels': wheels,
    'mpg': mpg,
    'vin': vin,
})

    
```
