```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Import Splinter and set the chromedriver path
from webdriver_manager.chrome import ChromeDriverManager
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Save KBB URL and visit the page
url = 'https://www.kbb.com/suv/'
browser.visit(url)

# Create BeautifulSoup object; parse with 'html.parser'
html = browser.html
soup = bs(html, "html.parser")
```

```python
results_view = soup.find('div', class_="css-zzr0gm-GridContainer ef2btag0")# 
print(results_view.prettify())
```

```python
for vehicle in results_view.find_all('h2'):
    print(vehicle.get_text())
    
    #vehicle.get_text(), append 
# browser.links.find_by_partial_text('next').click()
```
