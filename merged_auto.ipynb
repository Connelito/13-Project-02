```python
import pandas as pd
```

# AutoListData

```python
path = ('CleanResources/AutoListData.csv')
auto_df= pd.read_csv(path)
```

```python
auto_df.head()
```

```python
# auto_df['price']=auto_df['price'].str.replace(',','').str.replace('$','').astype(int)
auto_df['model']=auto_df['model'].str.replace('-','')
auto_df['mileage']=auto_df['mileage'].fillna(0)
auto_df['mileage']=auto_df['mileage'].str.replace('New','0').str.replace(',','').str.replace('Miles','')
auto_df['model']=auto_df['model'].str.replace('Pickup 1500','1500')
auto_df['price']=auto_df['price'].str.replace('$','').str.replace(',','')
auto_df
```

```python
auto_df=auto_df.sort_values(['year'])
auto_df
```

# USA_cars_datasets

```python
path2 = 'OGResources/USA_cars_datasets.csv'
auto2_df = pd.read_csv(path2)
```

```python
auto2_df
```

```python
auto2_df.isnull().sum()
```

```python
auto3_df=auto2_df.dropna()
auto3_df
```

```python
auto3_df['price'].dtype
```

```python
auto3_df=auto3_df.sort_values(['year'])
auto3_df
```

```python
auto3_df['price']=auto3_df['price'].astype(int)
auto3_df['year']=auto3_df['year'].astype(int)
auto3_df['mileage']=auto3_df['mileage'].astype(int)
auto3_df['lot']=auto3_df['lot'].astype(int)
auto3_df
```

```python
# Dropping my duplicated data and checking that the data has been shortened.
auto3_df.drop(auto3_df[auto3_df['model']=='door'].index,inplace=True)
# auto3_df.drop(auto3_df[auto3_df['price']=='0.0'].index,inplace=True)
# auto3_df.sort_values(['price'])
auto3_df=auto3_df.sort_values(['year'])
auto3_df
```

```python
auto3_df['price'].dtype
```

```python
auto3_df.drop(auto3_df[auto3_df['price']==0].index,inplace=True)
auto3_df
```

```python
auto3_df
```

```python
auto3_df.drop(auto3_df[auto3_df['year']<=1994].index,inplace=True)
auto3_df
```

```python
auto3_clean_df = auto3_df.rename(columns={
    'vin':'vin',
    'color':'display_color',
    'year':'year',
    'brand':'make',
    'model':'model',
    'price':'price',
    'mileage':'mileage',
    'lot':'dealer_name',
    'state':'state',
})
auto3_clean_df
```

```python
auto3_clean_df['trim']='N/A'
auto3_clean_df['body_style']='N/A'
auto3_clean_df['condition']='used'
auto3_clean_df['city']='N/A'
auto3_clean_df=auto3_clean_df.drop(['country'],axis=1)
auto3_clean_df
```

```python
autolist_df = pd.concat([auto3_clean_df, auto_df])
autolist_df
```

```python
autolist_df['state'] = autolist_df['state'].str.replace('alabama', 'AL').str.replace('alaska', 'AK').str.replace('arizona', 'AZ').str.replace('arkansas', 'AR').str.replace('california', 'CA').str.replace('colorado', 'CO').str.replace('connecticut', 'CT').str.replace('delaware', 'DE').str.replace('florida', 'FL').str.replace('georgia', 'GA').str.replace('hawaii', 'HI').str.replace('idaho', 'ID').str.replace('illinois', 'IL').str.replace('indiana', 'IN').str.replace('iowa', 'IA').str.replace('kansas', 'KS').str.replace('kentucky', 'KY').str.replace('louisiana', 'LA').str.replace('maine', 'ME').str.replace('maryland', 'MD').str.replace('massachusetts', 'MA').str.replace('michigan', 'MI').str.replace('minnesota', 'MN').str.replace('mississippi', 'MS').str.replace(
    'missouri', 'MO').str.replace('montana', 'MT').str.replace('nebraska', 'NE').str.replace('nevada', 'NV').str.replace('new Hampshire', 'NH').str.replace('new Jersey', 'NJ').str.replace('new Mexico', 'NM').str.replace('new York', 'NY').str.replace('north Carolina', 'NC').str.replace('north Dakota', 'ND').str.replace('ohio', 'OH').str.replace('oklahoma', 'OK').str.replace('oregon', 'OR').str.replace('pennsylvania', 'PA').str.replace('rhode Island', 'RI').str.replace('south Carolina', 'SC').str.replace('south Dakota', 'SD').str.replace('tennessee', 'TN').str.replace('texas', 'TX').str.replace('utah', 'UT').str.replace('vermont', 'VT').str.replace('virginia', 'VA').str.replace('washington', 'WA').str.replace('west Virginia', 'WV').str.replace('wisconsin', 'WI').str.replace('wyoming', 'WY')
autolist_df  #.reset_index(inplace=True)
```

```python
clean_autolist_df=autolist_df #.drop(columns='level_0').drop(columns='index')
clean_autolist_df
```

```python
clean_autolist_df.drop(clean_autolist_df[clean_autolist_df['model']=='doors'].index,inplace=True)
clean_autolist_df=clean_autolist_df.sort_values(['year'])
clean_autolist_df
```

```python
clean_autolist_df['model'].replace({'pk':'truck', 'pickup':'truck'}, inplace=True)
clean_autolist_df
```

```python
#write Merged DF to CSV
clean_autolist_df.to_csv('CleanResources/CleanAutoListData.csv', index=False)
```
